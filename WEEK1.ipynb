# Step 1 and 2

```python
import numpy as np
import pandas as pd
import cv2 as cv
from google.colab.patches import cv2_imshow
from skimage import io
from skimage import transform
from PIL import Image
import matplotlib.pylab as plt

urls = ["https://i.pinimg.com/564x/df/a4/16/dfa41669f4a334e0e181192a2add0564.jpg",
        "https://i.pinimg.com/564x/1f/eb/fe/1febfe330c948c908b83b9dff88d3a2d.jpg",
        "https://i.pinimg.com/564x/07/33/ba/0733ba760b29378474dea0fdbcb97107.jpg",
        "https://i.pinimg.com/736x/75/47/96/754796aee83e3a925a80abe18bb478a9.jpg"]


for url in urls:
  image = io.imread(url)
  image = cv.resize(image, (0,0), fx=0.5, fy=0.5)
  image_2 = cv.cvtColor(image, cv.COLOR_BGR2RGB)
  final_frame = cv.hconcat((image, image_2))
  cv2_imshow(final_frame)
  print('\n')

tinggi = image_2.shape[0]
lebar = image_2.shape[1]
print("Resolusi image: tinggi x lebar = ", tinggi, " x ", lebar)
cv2_imshow(image_2)

image_2 = cv.cvtColor(image, cv.COLOR_BGR2RGB)
image_3 = cv.cvtColor(image, cv.COLOR_BGR2RGB)

for y in range (lebar):
  image_3[int((tinggi)/2), y] = [255,255,255]

final_frame = cv.hconcat((image_2, image_3))
cv2_imshow(final_frame)
```

# TASK 2

```python
for x in range (tinggi):
  image_3[x, int((lebar)/2)] = [255,255,255]

final_frame = cv.hconcat((image_2, image_3))
cv2_imshow(final_frame)
```

```python
image_5 = cv.cvtColor(image, cv.COLOR_BGR2RGB)
for i in range(min(tinggi, lebar)):
    image_5[i][int(i*(lebar/tinggi))] = [255, 255, 255]
    image_5[tinggi - i - 1][int(i*(lebar/tinggi))] = [255, 255, 255]

final_frame = cv.hconcat((image_2, image_5))
cv2_imshow(final_frame)
```

# Task 3

```python
image_4 = cv.cvtColor(image, cv.COLOR_BGR2RGB)
line_length = 200

start_x = max(0, (lebar - line_length) // 2)
end_x = min(lebar, start_x + line_length)

middle_y = tinggi // 2

for x in range(start_x, end_x):
    image_4[middle_y, x] = [255, 255, 255]

final_frame = cv.hconcat((image_2, image_4))
cv2_imshow(final_frame)
```

# Task 4

```python
x_start, y_start = 50, 50
width, height = 100, 100

image_with_box = image_2.copy()
image_with_box[y_start:y_start+height, x_start:x_start+width] = [255, 255, 255]

final_frame = cv.hconcat((image_2, image_with_box))
cv2_imshow(final_frame)
```
