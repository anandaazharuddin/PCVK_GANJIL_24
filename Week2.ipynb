```python
black_img = np.zeros(shape=(512,512,3),dtype=np.int16)
plt.imshow(black_img)
```

```python
#add persegi
cv.rectangle(black_img,pt1=(200,200),pt2=(300,300),color=(0,0,255),thickness=15)
plt.imshow(black_img)
```

```python
#add line
cv.line(black_img,pt1=(0,0),pt2=(512,512),color=(255,255,255),thickness=5)
plt.imshow(black_img)
```

```python
#add circle
cv.circle(black_img,center=(100,100),radius=50,color=(255,0,0),thickness=8)
plt.imshow(black_img)
```

```python
#add text
font = cv.FONT_HERSHEY_SIMPLEX
cv.putText(black_img,text='Hello',org=(10,500),fontFace=font,fontScale=4,color=(255,255,0),thickness=2,lineType=cv.LINE_AA)
plt.imshow(black_img)
```

```python
import numpy as np
import cv2 as cv
import matplotlib.pyplot as plt

# Misalnya, kita membuat beberapa vertex (titik) sebagai contoh
vertices = np.array([[100, 50], [200, 300], [700, 200], [500, 100]], np.int32)

# Mengubah bentuk array sesuai dengan kebutuhan
pts = vertices.reshape((-1, 1, 2))

# Membuat gambar hitam (black image)
black_img = np.zeros((400, 800, 3), dtype=np.uint8)

# Menambahkan polyline pada gambar
cv.polylines(black_img, [pts], isClosed=True, color=(255, 0, 0), thickness=5)

# Menampilkan gambar
plt.imshow(black_img)
plt.show()
```
